## Data Sorting, Indexing and Grouping

### Sorting multiple values in descending order

```python
df.sort_values(["alcohol", "quality"], ascending = [False, True])
```

### Subsetting based on multiple conditions 

```python
df[df["alcohol"] > 11.0]
```

```python
df[df["quality"] > 7.0]
```

```python
df[df["residual sugar"] > 10.0]
```

## Data Cleaning - Replacing missing values and filling with the new ones

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.cleaned=df.dropna(axis=0)
df.cleaned
```

```python
df.filled=df.fillna(df.bfill())
df.filled
```

```python
df.filled.info()
```

### Looping and iterrows 

```python
for lab, row in df.iterrows():
    print (lab)
    print(row)
```

### Grouping 

```python
df.groupby("alcohol")["residual sugar"].count().sort_values(ascending=False).head(30)
```

## Data Visualisation 

```python
sns.relplot(x="alcohol", 
            y="residual sugar", 
            data = df,
            col="quality",
            col_wrap=2,
            kind="scatter",
            hue="quality")

plt.show()
```

```python
sns.scatterplot(x="alcohol", 
            y="pH", 
            data = df,
            hue="residual sugar")
plt.show()
```

```python
sns.scatterplot(x="alcohol", 
            y="residual sugar", 
            data = df,
            hue="quality")
plt.show()
```

```python
sns.scatterplot(x="alcohol", 
            y="fixed acidity", 
            data = df,
            size ="quality",
            hue="quality")
plt.show()
```

```python
sns.stripplot(x = "quality", y="total sulfur dioxide", data=df)
```
