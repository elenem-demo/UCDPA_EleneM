# Impoting Data from Kaggle

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_csv("WineQT.csv")
```

```python
df
```

## Data Sorting, Indexing and Grouping

### Sorting multiple values in descending order

```python
df.sort_values(["alcohol", "quality"], ascending = [False, True])
```

### Subsetting based on multiple conditions 

```python
df[df["alcohol"] > 11.0]
```

### Slicing and .loc

```python
df.loc[:, ["alcohol", "quality"]]
```

```python
df.iloc[[1,2,3]]
```

### Indexing

```python
df.columns
```

## Data Cleaning - Replacing missing values and filling with the new ones

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.cleaned=df.dropna(axis=0)
df.cleaned
```

```python
df.filled=df.fillna(df.bfill())
df.filled
```

```python
df.filled.info()
```

### Looping and iterrows 

```python
for lab, row in df.iterrows():
    print (lab)
    print(row)
```

### Grouping 

```python
df.groupby("alcohol")["residual sugar"].count().sort_values(ascending=False).head(30)
```

## Data Visualisation 

```python
sns.relplot(x="alcohol", 
            y="residual sugar", 
            data = df,
            kind="scatter")
plt.show()
```
